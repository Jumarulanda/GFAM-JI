```python
import numpy as np
import rabi_JC as rj
import matplotlib.pyplot as plt

from scipy.special import factorial as fac

hbar,m_e,e = 1,1,1

%matplotlib inline
plt.rcParams["font.family"] = "serif"
plt.rcParams['mathtext.fontset'] = 'dejavuserif'
plt.rcParams['mathtext.rm'] = 'dejavuserif'
plt.rcParams['mathtext.it'] = 'dejavuserif'
plt.rcParams['mathtext.bf'] = 'dejavuserif'
```

```python
c_eg = (1,0)
det = 0
n_bar = 5
alpha = np.sqrt(n_bar)
lamb = 1

r,theta = 0.001,np.pi/6

N = rj.N_sq(r,theta,alpha)

c_1 = rj.coherent_states
c_2 = rj.squeezed_states
c_3 = rj.squeezed_coherent_states
```

```python
t_range = np.linspace(0,30,1000)

fig,ax = plt.subplots(figsize=(20,10))

ax.plot(t_range, rj.W(t_range,lamb,det,c_eg,c_1,20,alpha))
ax.plot(t_range, rj.psi_mod(t_range,lamb,det,c_eg,c_1,20,alpha))
ax.axhline(y=0)
```

```python
fig,ax = plt.subplots(figsize=(20,10))

ax.plot(t_range, rj.W(t_range,lamb,det,c_eg,c_2,800,r,theta))
ax.plot(t_range, rj.psi_mod(t_range,lamb,det,c_eg,c_2,800,r,theta))
ax.axhline(y=0)
```
