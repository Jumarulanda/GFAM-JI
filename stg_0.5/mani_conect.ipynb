```python
import numpy as np
import rabi_JC as rj
import rabi_box as rb
import matplotlib.pyplot as plt

from scipy.special import factorial as fac

hbar,m_e,e = 1,1,1

%matplotlib inline
plt.rcParams["font.family"] = "serif"
plt.rcParams['mathtext.fontset'] = 'dejavuserif'
plt.rcParams['mathtext.rm'] = 'dejavuserif'
plt.rcParams['mathtext.it'] = 'dejavuserif'
plt.rcParams['mathtext.bf'] = 'dejavuserif'
```

```python
def tras_1(omeg,det):
    return 2*det**2 / omeg**2

def tras_2(omeg,det,n,omeg_n,det_n):
    t1 = np.sin(omeg_n/omeg * (2*n+1)*np.pi/2)**2 * 2 * det_n**2 / omeg_n**2
    t2 = np.cos(omeg_n/omeg * (2*n+1)*np.pi)
    
    return t1+t2+(-1)**(n+1)

#for det=0

def conect(n_huh,n_hav,lamb):
    mt = (n_hav + 1) / (n_huh + 1)
    return np.sqrt(mt)*lamb
```

```python
c_eg = (1,0)
det_1,det_2 = 0,0

n_1,n_2 = 0,1

c_n1 = lambda n: int((n==n_1))
c_n2 = lambda n: int((n==n_2))

lamb_1 = 0.1

omeg_1 = rb.omega_q(det_1,lamb_1,n_1)
omeg_1
```

```python
omeg_2 = np.linspace(-0.1,0.1,1000)

plt.plot(omeg_2,tras_1(omeg_2,det_2))
plt.plot(omeg_2,tras_2(omeg_2,det_2,0,omeg_1,det_1))

plt.ylim(-0.02,0.02)

lamb_2 = conect(n_2,n_1,lamb_1)
lamb_2
```
