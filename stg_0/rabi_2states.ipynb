```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.constants import hbar

%matplotlib inline
plt.rcParams["font.family"] = "serif"
cmap = plt.get_cmap('gnuplot')

plt.rcParams['mathtext.fontset'] = 'dejavuserif'
plt.rcParams['mathtext.rm'] = 'dejavuserif'
plt.rcParams['mathtext.it'] = 'dejavuserif'
plt.rcParams['mathtext.bf'] = 'dejavuserif'
```

```python
# constants

E0,L = 1,1

d = 9*np.pi**3 * hbar / 32

alpha = -d*E0*8*L**2 / 9 / np.pi**2

omega = lambda det: np.sqrt(det**2 + alpha**2 / hbar**2)
```

```python
def prob_1(t,det):
    p1_1 = np.cos(omega(det)*t/2)**2
    p1_2 = np.sin(omega(det)*t/2)**2 * det**2 / omega(det)**2
    
    return p1_1 + p1_2

def prob_2(t,det):
    
    return np.sin(omega(det)*t/2)**2 * alpha**2 / (hbar**2 * omega(det)**2)
```

```python
t = np.linspace(0,15,100)

plt.plot(t,prob_1(t,0),linestyle='-',color='#2a2ef6',linewidth=1)
plt.plot(t,prob_1(t,1),linestyle='-',color='#e91717',linewidth=1)
plt.plot(t,prob_1(t,2),linestyle='-',color='#48ef06',linewidth=1)
```
