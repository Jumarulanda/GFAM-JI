```python
alpha = np.sqrt(n_bar)
lamb = 1

c_n = rb.coherent_states

def coef_g(t):
    par_sum = [abs(c_n(0,alpha))**2 * np.sin(lamb*np.sqrt(0+1)*t)**2]
    for n in range(1,100):
        par_sum.append(par_sum[-1] + abs(c_n(n,alpha))**2 * np.sin(lamb*np.sqrt(n+1)*t)**2)
    return par_sum

def coef_e(t):
    par_sum = [abs(c_n(0,alpha))**2 * np.sin(lamb*t)**2]
    for n in range(1,100):
        par_sum.append(par_sum[-1] + abs(c_n(n,alpha))**2 * np.cos(lamb*np.sqrt(n+1)*t)**2)
    return par_sum

def coef_ne(t):
    par_sum = [abs(c_n(0,alpha))**2 * np.cos(lamb*np.sqrt(1)*t)**2]
    for n in range(1,100):
        par_sum.append(par_sum[-1] + abs(c_n(n,alpha))**2 * np.cos(lamb*np.sqrt(n+1)*t)**2 )
    return par_sum
    
def coef_ng(t):
    par_sum = [abs(c_n(0,alpha))**2 * np.sin(lamb*np.sqrt(1)*t)**2]
    for n in range(2,100):
        par_sum.append(par_sum[-1] + abs(c_n(n-1,alpha))**2 * np.sin(lamb*np.sqrt(n)*t)**2 )
    return par_sum
```

```python
t = 2

ps1 = coef_g(t)
ps2 = coef_e(t)

ps3 = coef_ng(t)
ps4 = coef_ne(t)

fig,[ax1,ax2] = plt.subplots(nrows=1,ncols=2,figsize=(10,4))

ax1.plot(ps1)
ax1.plot(ps2)
ax1.plot([g+e for g,e in zip(ps1,ps2)])
ax1.set_ylim(0,1.2)

ax2.plot(ps3)
ax2.plot(ps4)
ax2.plot([g+e for g,e in zip(ps3,ps4)])
ax2.set_ylim(0,1.2)
```
